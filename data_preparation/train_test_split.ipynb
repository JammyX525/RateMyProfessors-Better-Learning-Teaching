```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('./RMP_cleaned.csv')
df.info()
```

```python
df_modified = df.drop(['Name', 'OverallRate', 'TakeAgain', 'Overall_Difficulty', 'Tags', 'Emotion'], axis = 1)
df_modified
```

```python
df_modified = df_modified.rename(columns={"StudentDifficulty": "Difficulty",
                                                                        "Student_TakeAgain": "TakeAgain"})
```

```python
df_modified
```

```python
df_modified['GoodTeaching'] = (df_modified['QualityRate'] > 3.0).astype('int64')
df_modified
```

```python
df_modified.drop(['QualityRate'], axis = 1, inplace=True)
df_modified
```

```python
list(df_modified.columns)
```

```python
col_names = list(df_modified.columns)
col_names.remove('GoodTeaching')
col_names.insert(0, 'GoodTeaching')
col_names
```

```python
df_reindexed = df_modified[col_names]
df_reindexed
```

```python
df_reindexed.info()
```

```python
comment_df = df_reindexed[['GoodTeaching', 'Comment']]
comment_df
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(df_reindexed.iloc[:, 1:], df_reindexed['GoodTeaching'], 
                                                    test_size=0.25, stratify=df_reindexed['GoodTeaching'], random_state=24)
```

```python
X_train.shape, X_test.shape
```

```python
train_df = pd.concat([y_train, X_train], axis = 1)
test_df = pd.concat([y_test, X_test], axis = 1)
```

```python
comment_train = train_df[['GoodTeaching', 'Comment']]
comment_test = test_df[['GoodTeaching', 'Comment']]
```

```python
train_df.drop(['Comment'], axis = 1, inplace=True)
test_df.drop(['Comment'], axis = 1, inplace=True)
```

```python
train_df.to_csv('./rmp_train.csv', index=False)
test_df.to_csv('./rmp_test.csv', index=False)

comment_train.to_csv('./comment_train.csv', index=False)
comment_test.to_csv('./comment_test.csv', index=False)
```
